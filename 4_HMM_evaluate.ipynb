```python
from HMM import *
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
T = np.array([[0.3 ,0.6,0.1],[0.3,0.4,0.3],[0.5,0.1,0.4]])  # each row gives transition probability for a state
E = np.array([[0.25,0.45,0.30],[0.45,0.35,0.30],[0.15,0.55,0.30]])  # each row gives emission probability for a state
W = np.array(['W1','W2','W3'])
V = np.array(['V1','V2','V3'])

print('Transition probabiity')
print(T)
print()
print('Emission probabiity')
print(E)
```

## Evaluation

- Given $(\theta,V^T)$ estimate $P(V^T / \theta)$ 

```python
seq = np.random.randint(2,size=(1,50))

initial_proba = np.array([0.4,0.3,0.3])

print('Given emitted sequence :',seq)
```

```python
A1 = HMM_evaluate(T,E,W,V)

A1.evaluate_sequence(seq,initial_proba)  

# gives the probability of obtaining the sequence given the HMM
```

```python
fmatrix = A1.forward_algorithm_matrix(seq,initial_proba)
```

```python
plt.figure(figsize=(30,7))
sns.heatmap(np.log(fmatrix),annot=True)  # log transform
plt.xlabel('Timestep')
plt.ylabel('Hidden state')
plt.yticks([0,1,2],['$W_0$','$W_1$','$W_2$'])
plt.show()
```
